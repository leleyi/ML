```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
import mglearn
```

#### Data processing

```python
names=['Class', 'age','menopause', 'tumor-size','inv-nodes', 'node-caps','deg-malig', 'breast', 'breast-quad', 'irradiat']
df = pd.read_csv("./breast-cancer.data", header=0, names = names)
df.head()
```

```python
# no-recurrence-events 0 , recurrence-events 1 label to num
df.iloc[:,0] = LabelEncoder().fit_transform(df.iloc[:,-1]) 
# attribute to num 
df.iloc[:,1:] = OrdinalEncoder().fit_transform(df.iloc[:,1:])
X_train,X_test,y_train,y_test = train_test_split(df.values[:,1:-1],df.values[:,0],random_state=0)
```

#####  Draw histograms for features

```python
df.hist(bins = 100,figsize  = (16, 12))
# print(df.info)
print("Investigate features of your variant \nDraw histograms for features")
plt.show()
```

##### For a classification task draw scatter plot matrix of features.

```python
colors=['red','green']

# make a scatter plot
pd.plotting.scatter_matrix(df,figsize=[20,20],marker='x',c=df.Class.apply(lambda x:colors[x]),range_padding=2)
plt.show()
```

#####   Correlation coefficient

* **what features are most useful for solving the problem and why** :
    - so the irradiat is the most useful feature for solving the problem, beacase the correlation coefficient is the biggest; 

```python
corr_matrix = df.corr()
corr_matrix["Class"].sort_values(ascending=False)
```

```python
corr_matrix["irradiat"].sort_values(ascending=False)
```

#### Train
##### Estimate your the result of training

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier().fit(X_train, y_train)
fig, ax = plt.subplots(figsize=(20, 20)) 
plot_tree(dt,filled=True)
plt.show()
```

```python
#let us get the predictions using the classifier we had fit above
#import confusion_matrix
from sklearn.metrics import confusion_matrix
y_pred = dt.predict(X_test)
print(y_pred)
confusion_matrix(y_test,y_pred)
```
